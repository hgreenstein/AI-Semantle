```python
import gensim
import numpy as np
#import random
from sklearn.cluster import KMeans
#import tkinter as tk
from answers import secretWords as answers
from constants import PATH_TO_DATASET

gnews_model = gensim.models.KeyedVectors.load_word2vec_format(PATH_TO_DATASET, binary=True)
```

```python
embeddings = []
for answer in answers:
    embeddings.append(gnews_model[answer])
kModel = KMeans(n_clusters=8, random_state=0, n_init=10)
clusters = kModel.fit_predict(embeddings)
```

```python
import plotly.express as px
import umap
```

```python
reducer = umap.UMAP(random_state=42)
embeddings_2d = reducer.fit_transform(embeddings)
import pandas as pd

df = pd.DataFrame(embeddings_2d, columns=["x", "y"])
df["cluster"] = clusters
df["word"] = answers
fig = px.scatter(
    df,
    x="x",
    y="y",
    color="cluster",
    text="word",
    hover_data=["word"],
    color_continuous_scale=px.colors.qualitative.Plotly,
    title="Word Clusters",
    width=1200,
    height=800,
)

# Improve text visibility and update the color scale
fig.update_traces(
    textposition="top center",
    textfont_size=10,
    marker=dict(size=10, line=dict(width=2, color="black")),
)
fig.update_layout(showlegend=False)

fig.show()

```
