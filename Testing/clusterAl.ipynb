```python
import gensim
import json
import numpy as np
import plotly.express as px
from sklearn.cluster import AgglomerativeClustering
from answers import secretWords as answers
import umap
from constants import PATH_TO_DATASET

# Load pre-trained word embeddings
gnews_model = gensim.models.KeyedVectors.load_word2vec_format(PATH_TO_DATASET, binary=True)

# Get embeddings for answer words
embeddings = []
for answer in answers:
    embeddings.append(gnews_model[answer])

```

```python
n = 4
agg_clustering = AgglomerativeClustering(n_clusters=n, metric='cosine', linkage='complete')
cluster_labels = agg_clustering.fit_predict(embeddings)
```

```python
clusterWords = [[], [], [], []]
for i, j in enumerate(cluster_labels):
    clusterWords[j - 1].append(answers[i])
with open('clusters.py', 'w') as f:
    f.write('clusters = ')
    json.dump(clusterWords, f)
embeddings_zero = []
embeddings_one = []
for word in clusterWords[0]:
    embeddings_zero.append(gnews_model[word])
cluster_labels_zero = agg_clustering.fit_predict(embeddings_zero)
clusterWordsZero = [[], [], [], []]
for i, j in enumerate(cluster_labels_zero):
    clusterWordsZero[j - 1].append(clusterWords[0][i])
for i, j in enumerate(clusterWordsZero):
    print(i, len(clusterWordsZero[i]))
for word in clusterWords[1]:
    embeddings_one.append(gnews_model[word])
cluster_labels_one = agg_clustering.fit_predict(embeddings_one)
clusterWordsOne = [[], [], [], []]
for i, j in enumerate(cluster_labels_one):
    clusterWordsOne[j - 1].append(clusterWords[1][i])
for i, j in enumerate(clusterWordsOne):
    print(i, len(clusterWordsOne[i]))
embeddings_zero_zero = []
for word in clusterWordsZero[0]:
    embeddings_zero_zero.append(gnews_model[word]) 
cluster_labels_zero_zero = agg_clustering.fit_predict(embeddings_zero_zero)
clusterWordsZeroZero = [[], [], [], []]
for i, j in enumerate(cluster_labels_zero_zero):
    clusterWordsZeroZero[j - 1].append(clusterWordsZero[0][i])
for i, j in enumerate(clusterWordsZeroZero):
    print(i, len(clusterWordsZeroZero[i]))
    

```

```python
# Perform hierarchical clustering
n = 8
agg_clustering = AgglomerativeClustering(n_clusters=n, metric='cosine', linkage='complete')
cluster_labels = agg_clustering.fit_predict(embeddings)
```

```python
clusterWords = [[], [], [], [], [], [], [], []]
for i, j in enumerate(cluster_labels):
    clusterWords[j - 1].append(answers[i])
with open('clusters.py', 'w') as f:
    f.write('clusters = ')
    json.dump(clusterWords, f)
for i, j in enumerate(clusterWords):
    print(i, len(clusterWords[i]))
embeddings_zero = []
embeddings_one = []
for word in clusterWords[0]:
    embeddings_zero.append(gnews_model[word])
cluster_labels_zero = agg_clustering.fit_predict(embeddings_zero)
clusterWordsZero = [[], [], [], [], [], [], [], []]
for i, j in enumerate(cluster_labels_zero):
    clusterWordsZero[j - 1].append(clusterWords[0][i])
for i, j in enumerate(clusterWordsZero):
    print(i, len(clusterWordsZero[i]))
for word in clusterWords[1]:
    embeddings_one.append(gnews_model[word])
cluster_labels_one = agg_clustering.fit_predict(embeddings_one)
clusterWordsOne = [[], [], [], [], [], [], [], []]
for i, j in enumerate(cluster_labels_one):
    clusterWordsOne[j - 1].append(clusterWords[1][i])
for i, j in enumerate(clusterWordsOne):
    print(i, len(clusterWordsOne[i]))
embeddings_zero_zero = []
for word in clusterWordsZero[0]:
    embeddings_zero_zero.append(gnews_model[word]) 
cluster_labels_zero_zero = agg_clustering.fit_predict(embeddings_zero_zero)
clusterWordsZeroZero = [[], [], [], [], [], [], [], []]
for i, j in enumerate(cluster_labels_zero_zero):
    clusterWordsZeroZero[j - 1].append(clusterWordsZero[0][i])
for i, j in enumerate(clusterWordsZeroZero):
    print(i, len(clusterWordsZeroZero[i]))
    

```

```python
# Reduce the dimensionality of embeddings using UMAP
reducer = umap.UMAP()
reduced_embeddings = reducer.fit_transform(embeddings)
data = {
    'UMAP 1st Component': reduced_embeddings[:, 0],
    'UMAP 2nd Component': reduced_embeddings[:, 1],
    'Word': answers,
    'Cluster': cluster_labels
}
fig = px.scatter(data_frame=data, x='UMAP 1st Component', y='UMAP 2nd Component', color='Cluster', text='Word')
fig.update_traces(textposition='top center', textfont_size=8)
fig.update_layout(title='Hierarchical Clustering Visualization', legend_title_text='Cluster')
fig.show()
```
